```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("normalized_topics_and_clusters_correct.xlsx")
sns.set(style="whitegrid")
```

```python
df.columns
```

```python
# Cluster mapping 
cluster_map = {
    "Terrorism, Paramilitary Violence, and Security Cooperation": [
        'Cross-Border Legal Frameworks',
        'Paramilitary Activities & Ceasefires',
        'Political Dialogue & Prisoner Releases',
    ],
    "Peace Agreements and Political Settlements": [
        'Agreement Negotiation & Implementation',
        'Security Reforms (Decommissioning & Policing)',
        'Good Friday Agreement Protections',
        'Human Rights & Prisoner Policies'
    ],
    "UK/Ireland & Cross-Border Relations": [
        'Anglo-Irish Agreement',
        'Governance UK/Ireland',
        'North-South Coordination'
    ]
}

cluster_names = list(cluster_map.keys())
```

```python
plt.figure(figsize=(12, 6))
for cluster in cluster_names:
    sns.lineplot(data=df, x='year', y=cluster, label=cluster)

plt.xlabel("Year")
plt.ylabel("Normalized Probability")
plt.title(" Cluster Trends Over Time (Normalized)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python

for cluster in cluster_names:
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='year', y=cluster, marker='o', color='steelblue')
    plt.xlabel('Year')
    plt.ylabel('Normalized Cluster Probability')
    plt.title(f'{cluster} (Aggregated) Over Time')
    plt.tight_layout()
    plt.show()
```

```python
for cluster_name, topics in cluster_map.items():
    plt.figure(figsize=(12, 6))
    for topic in topics:
        sns.lineplot(data=df, x='year', y=topic, label=topic)
    
    plt.xlabel("Year")
    plt.ylabel("Normalized Probability")
    plt.title(f"{cluster_name}  Topic-Level Trends")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
